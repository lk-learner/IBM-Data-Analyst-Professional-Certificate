#### Exercise 1: 
Read Total profit of all months and show it using a line plot
Total profit data provided for each month. Generated line plot must include the following properties: –

X label name = Month Number

Y label name = Total profit


```python
# Extract the Month Number and Total Profit columns
month_profit = df[['month_number', 'total_profit']]

# Create a line plot
plt.plot(month_profit['month_number'], month_profit['total_profit'])

# Set the x and y axis labels
plt.xlabel('Month Number')
plt.ylabel('Total profit')

# Show the plot
plt.show()
```

#### Exercise 2: 
Get total profit of all months and show line plot with the following Style properties
Generated line plot must include following Style properties: –

	Line Style dotted and Line-color should be red

Show legend at the lower right location.

X label name = Month Number

Y label name = Sold units number

Add a circle marker.

Line marker color as read

Line width should be 3


```python
# Get the total profit of all months
month_profit = df.groupby(['month_number'])['total_profit'].sum().reset_index()

# Create a line plot with dotted line style, red line color, circle marker, and 3 line width
plt.plot(month_profit['month_number'], month_profit['total_profit'], linestyle='dotted', color='red', marker='o', markersize=8, markerfacecolor='red', linewidth=3)

# Add legend at the lower right location
plt.legend(['Total Profit'], loc='lower right')

# Set the x and y axis labels
plt.xlabel('Month Number')
plt.ylabel('Units Sold')

# Show the plot
plt.show()
```

#### Exercise 3: 
Read all product sales data and show it using a multiline plot

Display the number of units sold per month for each product using multiline plots. (i.e., Separate Plotline for each product).


```python
# Create a multiline plot with a separate plot line for each column
columns = ['facecream', 'facewash', 'toothpaste', 'bathingsoap', 'shampoo', 'moisturizer']
for column in columns:
    plt.plot(df['month_number'], df[column], label=column)

# Add legend at the lower right location
plt.legend(loc='lower right')

# Set the x and y axis labels
plt.xlabel('Month Number')
plt.ylabel('Units sold')

# Show the plot
plt.show()
```

#### Exercise 4: 
Read toothpaste sales data of each month and show it using a scatter plot

Also, add a grid in the plot. gridline style should “–“.


```python
# Create a scatter plot of toothpaste sales data
plt.scatter(df['month_number'], df['toothpaste'])

# Add gridlines with style "-"
plt.grid(linestyle="-")

# Set the x and y axis labels
plt.xlabel('Month Number')
plt.ylabel('Units sold')

# Show the plot
plt.show()
```

#### Exercise 5: 
Read face cream and facewash product sales data and show it using the bar chart

```python
# Create a bar chart of the total units sold for face cream and facewash
plt.bar(['facecream', 'facewash'], df['total_units'][0:2])

# Set the x and y axis labels
plt.xlabel('Product')
plt.ylabel('Units sold')

# Show the plot
plt.show()
```

#### Exercise 7: 
Read the total profit of each month and show it using the histogram to see the most common profit ranges

```python
# Calculate the total profit for each month
Total_Profit = df['total_profit']

# Create a histogram of the profit ranges
plt.hist(Total_Profit, bins=10)

# Set the x and y axis labels
plt.xlabel('Total Profit')
plt.ylabel('Frequency')

# Show the plot
plt.show()
```

#### Exercise 8: 
Calculate total sale data for last year for each product and show it using a Pie chart

```python
# Calculate total sales data for each month for each product
product_sales_data = df.groupby(['month_number']).sum()

# Create a pie chart for each product for each month
for product in product_sales_data.columns:
    plt.figure()
    plt.pie(product_sales_data[product], labels=product_sales_data.index)
    plt.title(f"{product} Sales by Month")
    plt.show()
```

#### Exercise 9: 
Read Bathing soap facewash of all months and display it using the Subplot

```python
# Read the sales data for 'bathingsoap' and 'facewash'
bathingsoap = df['bathingsoap']
facewash = df['facewash']

# Create a subplot with 2 rows and 1 column
fig, axs = plt.subplots(2, 1, figsize=(8,8))

# Plot the 'bathingsoap' sales data on the first subplot
axs[0].plot(bathingsoap)
axs[0].set_title('Bathing Soap Sales')

# Plot the 'facewash' sales data on the second subplot
axs[1].plot(facewash)
axs[1].set_title('Facewash Sales')

# Add x and y axis labels to each subplot
axs[0].set_ylabel('Units Sold')
axs[1].set_ylabel('Units Sold')
axs[1].set_xlabel('Month Number')

# Adjust the padding between subplots
plt.tight_layout()

# Show the plot
plt.show()
```

#### Exercise 10: 
Read all product sales data and show it using the stack plot

```python
# Read the product sales data for each month
sales_data = df[['facecream', 'facewash', 'toothpaste', 'bathingsoap', 'shampoo', 'moisturizer']].values

# Set the labels for each product
product_labels = ['Face Cream', 'Face Wash', 'Toothpaste', 'Bathing Soap', 'Shampoo', 'Moisturizer']

# Create the stack plot
plt.stackplot(df['month_number'], sales_data.T, labels=product_labels)

# Add the x and y axis labels and the legend
plt.xlabel('Month Number')
plt.ylabel('Total Units Sold')
plt.legend(loc='upper left')

# Show the plot
plt.show()
```
