#### 3. Find the Z-Score for column Recovered


```python
from scipy import stats

# Calculate the z-score for column "Recovered"
z_scores = stats.zscore(df["Recovered"])

# Print the resulting array of z-scores
print(z_scores)
```

#### 5. Write a Python program to get the maximum number of confirmed, deaths, recovered and active cases of Novel Coronavirus (COVID-19) Country wise.

```python
df['Country_Region'].unique()
```

```python
# Group the data by country and get the maximum value for the "confirmed" column
confirmed_max = df.groupby(['Country_Region'])['Confirmed'].max()
deaths_max = df.groupby(['Country_Region'])['Deaths'].max()
recovered_max = df.groupby(['Country_Region'])['Recovered'].max()

# Show the resulting dataframe
print(confirmed_max)
print(deaths_max)
print(recovered_max)
```

#### 6. Write a Python program to get the total number of confirmed deaths and recovered people of Novel Coronavirus (COVID-19) cases Country/Region - Province/State wise


```python
grouped_data = df.groupby(['Country_Region', 'Province_State'])[['Confirmed', 'Deaths','Recovered']].sum()

# Show the resulting dataframe
print(grouped_data)
```

#### 7. Write a Python program to create a plot (lines) of total deaths, confirmed, recovered and active cases Country wise

```python
import matplotlib.pyplot as plt
# Create a line plot of the data
df.plot(kind='line', figsize=(15,8), x='')

# Set the plot title and axes labels
plt.title('COVID-19 Cases by Country')
plt.xlabel('Country-Province')
plt.ylabel('Number of Cases')

# Show the plot
plt.show()
```

```python
import plotly.graph_objs as go
# Create a Plotly figure
fig = go.Figure()

# Add traces for each type of case
fig.add_trace(go.Scatter(x=df['Province_State'], y=df['Confirmed'], name='Confirmed'))
fig.add_trace(go.Scatter(x=df[ 'Province_State'], y=df['Recovered'], name='Recovered'))
fig.add_trace(go.Scatter(x=df[ 'Province_State'], y=df['Deaths'], name='Deaths'))
fig.add_trace(go.Scatter(x=df[ 'Province_State'], y=df['Active'], name='Active'))

# Update the layout
fig.update_layout(title='COVID-19 Cases by Country - Province',
                  xaxis_title='Country-Province',
                  yaxis_title='Number of Cases',
                  template='plotly_dark')

# Show the plot
fig.show()
```
